# Draw Bitcoin Price and the Elon Musk time of posting on the same plot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
btc_price = pd.read_csv('../../dataset/Bitcoin/daily/Bitcoin_2010_2024.csv')
elon_tweets = pd.read_csv('../../dataset/Tweet/TweetsElonMuskBitcoinFiltered.csv')
```

```python
btc_price['Start'] = pd.to_datetime(btc_price['Start'])
plt.figure(figsize=(15,10))
plt.plot(btc_price['Start'], btc_price['Open'], label='Bitcoin Price')
```

```python
btc_price['Start'] = pd.to_datetime(btc_price['Start'])
btc_price_2018_to_2021 = btc_price.loc[(btc_price['Start'] >= '2018-01-01') & (btc_price['Start'] <= '2021-12-31')]
plt.figure(figsize=(15,10))
plt.plot(btc_price_2018_to_2021['Start'], btc_price_2018_to_2021['Open'], label='Bitcoin Price')
plt.show()
```

```python
elon_tweets['date'] = pd.to_datetime(elon_tweets['date'],dayfirst=True)
plt.figure(figsize=(15,10))

plt.xticks(rotation=45)

plt.scatter(elon_tweets['date'],range(len(elon_tweets['date'])), label='Elon Musk Tweet', color='red')
# plt.plot(elon_tweets['date'], range(len(elon_tweets['date'])), label='Elon Musk Tweets')

plt.show()
```
